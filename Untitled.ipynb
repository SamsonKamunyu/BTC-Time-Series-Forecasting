## Time Series Modelling

```python
# Splitting the data into training and testing
train = data[data.index < pd.to_datetime('2021-12-31', format = '%Y-%m-%d')]
test = data[data.index >= pd.to_datetime('2021-12-31', format = '%Y-%m-%d')]

#Visualize the training and test data
ax, bar = plt.subplots(figsize = (12, 10))
bar = plt.plot(train, color = 'black')
bar = plt.plot(test, color = 'red')

# Label the axes
plt.xlabel('Date',  fontsize = 14)
plt.ylabel('BTC-USD Price', fontsize = 14)

# Label Title
plt.xlabel('Train/Test BTC-USD Price from Jan 2019 to Nov 2022', fontsize = 15)

# Display
plt.show()
```

## Autoregressive Moving Average

```python
# Import the ARMA
from statsmodels.tsa.statespace.sarimax import SARIMAX
```

```python
# Define the out input
y = train['BTC-USD']
```

```python
# Initialize the model
ARMA_model = SARIMAX(y, order = (1, 0, 1))
```

```python
# Fit the model
ARMA_model = ARMA_model.fit()
```

```python
# Make predictions on the data
y_pred = ARMA_model.get_forecast(len(test.index))
y_pred_df = y_pred.conf_int(alpha = 0.05)
y_pred_df['Predictions'] = ARMA_model.predict(start = y_pred_df.index[0], end = y_pred_df.index[-1])
y_pred_df.index = test.index
y_pred_output = y_pred_df['Predictions']
```

```python
#Visualize the predictions output
ax, bar = plt.subplots(figsize = (12, 10))
bar = plt.plot(train, color = 'black', label = 'Training')
bar = plt.plot(test, color = 'red', label = 'Testing')
bar = plt.plot(y_pred_output, color = 'green', label = 'ARMA Predictions')

# Label the axes
plt.xlabel('Date',  fontsize = 14)
plt.ylabel('BTC-USD Price', fontsize = 14)

# Label Title
plt.xlabel('Train/Test Split BTC-USD Price from Jan 2019 to Nov 2022', fontsize = 15)

# Legend
plt.legend()

# Display
plt.show()
```

```python
# Evaluations
arma_rmse = np.sqrt(mean_squared_error(test['BTC-USD'].values, y_pred_df['Predictions']))
print(f'RMSE: {arma_rmse}')
```

## Autoregressive Integrated Moving Average

```python
from statsmodels.tsa.arima.model import ARIMA
```

#### ARIMA has 3 important parameters to note:
- 1st parameter corresponds to lagging
* 2nd " " " differencing(non stationary data to stationary)
* 3rd " " " white noise

```python
# Initializing the ARIMA model
ARIMAmodel = ARIMA(y, order = (2, 1, 0))
```

```python
# fit the model
ARIMAmodel = ARIMAmodel.fit()
```

```python
# make predictions on the data 
y_pred_arima = ARIMAmodel.get_forecast(len(test.index))
y_pred_df_arima = y_pred_arima.conf_int(alpha = 0.05)
y_pred_df_arima["Predictions"] = ARIMAmodel.predict(start = y_pred_df_arima.index[0], end = y_pred_df_arima.index[-1])
y_pred_df_arima.index = test.index
y_pred_output_arima = y_pred_df_arima["Predictions"]
```

```python
# visualize the predictions output 
ax, bar = plt.subplots(figsize=(12, 10))
bar = plt.plot(train, color = "black", label = "Training")
bar = plt.plot(test, color = "red", label = "Testing")
bar = plt.plot(y_pred_output, color = "green", label = "ARMA Predictions")
bar = plt.plot(y_pred_output_arima, color = "yellow", label = "ARIMA Predictions")

# label the axes
plt.xlabel("Date", fontsize = 14)
plt.ylabel("BTC-USD Price", fontsize = 14)
# title
plt.title("Train/ Test Split BTC-USD Price From Jan 2019 to Nov 2022", fontsize=15)
# legend 
plt.legend()
# display 
plt.show()
```

```python
# evaluations 
arima_rmse = np.sqrt(mean_squared_error(test["BTC-USD"].values, 
                                        y_pred_df_arima["Predictions"]))
print(f"RMSE: {arima_rmse}")
```

## Seasonal ARIMA(SARIMA)

```python
# initialize SARIMA mode
SARIMAXmodel = SARIMAX(y, order = (2, 1, 0), seasonal_order = (2, 2, 2, 11))
```

```python
# fit the model
SARIMAXmodel = SARIMAXmodel.fit()
```

```python
# make predictions on the data 
y_pred_sarima = SARIMAXmodel.get_forecast(len(test.index))
y_pred_df_sarima = y_pred_sarima.conf_int(alpha=0.05)
y_pred_df_sarima["Predictions"] = SARIMAXmodel.predict(start = y_pred_df_sarima.index[0], end = y_pred_df_sarima.index[-1])
y_pred_df_sarima.index = test.index
y_pred_output_sarima = y_pred_df_sarima["Predictions"]
```

```python
# visualize the predictions output 
ax, bar = plt.subplots(figsize=(12, 10))
bar = plt.plot(train, color = "black", label = "Training")
bar = plt.plot(test, color = "red", label = "Testing")
bar = plt.plot(y_pred_output, color = "green", label = "ARMA Predictions")
bar = plt.plot(y_pred_output_arima, color = "yellow", label = "ARIMA Predictions")
bar = plt.plot(y_pred_output_sarima, color = "blue", label = "SARIMA Predictions")

# label the axes
plt.xlabel("Date", fontsize = 14)
plt.ylabel("BTC-USD Price", fontsize = 14)
# title
plt.title("Train/ Test Split BTC-USD Price From Jan 2019 to Nov 2022", fontsize=15)
# legend 
plt.legend()
# display 
plt.show()
```

```python
# evaluations 
sarima_rmse = np.sqrt(mean_squared_error(test["BTC-USD"].values, y_pred_df_sarima["Predictions"]))
print(f"RMSE: {sarima_rmse}")
```
